# Cleaning the data

```python
# checking for the null values
movie_df.isna().sum()
```

```python
movie_df.isna().sum()
```

```python
# dropping the missing rows in year
movie_df.dropna(subset=['Year'], inplace=True)
```

```python
movie_df['Year'] = movie_df['Year'].astype(str).str.extract('(\d+)').astype(float)
```

```python
movie_df.dtypes
```

```python
movie_df.head()
```

```python
movie_df.isna().sum()
```

```python
movie_df.shape
```

```python
movie_df['Votes'].unique()
```

```python
movie_df['Votes']=movie_df['Votes'].replace("$5.16M", 516)
```

```python
movie_df['Votes'] = pd.to_numeric(movie_df['Votes'].str.replace(',', ''))
```

```python
movie_df.isna().sum()
```

```python
# Remove 'min' from the 'Duration' column and convert to integers
movie_df['Duration'] = movie_df['Duration'].str.replace(' min', '').apply(pd.to_numeric)
```

```python
movie_df.dtypes
```

```python
# filling the missing rows in Actor with unknown value
movie_df.dropna(subset=['Director', 'Actor 1', 'Actor 2', 'Actor 3', 'Genre'],inplace=True)
```

```python
movie_df.isna().sum()
```

```python
movie_df.isna().sum()
```

```python
# Filling the missing Ratings values with mean
movie_df['Rating'].fillna(movie_df['Rating'].mean(), inplace=True)
```

```python
# Filling the missing duration value with average duration
movie_df['Duration'].fillna(movie_df['Duration'].median(), inplace=True)
```

```python
# Filling the votes with average votes 
movie_df['Votes'].fillna(movie_df['Votes'].median(), inplace=True)
```

# 4.  Data Encoding

```python
movie_df.dtypes
```

```python
# Encoding the categorical features using frequency encoding
cat_col=['Name', 'Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']

for col in cat_col:
    frequency_map = movie_df[col].value_counts(normalize=True).to_dict()
    movie_df[col + '_encoded'] = movie_df[col].map(frequency_map)

# dropping the original features
movie_df.drop(['Name', 'Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3'], axis=1, inplace=True)
```

# 4 Modeling and evaluation

first step of modeling is to split the data into train and test sets

```python
# Split the data into X and y
X=movie_df.drop("Rating",axis=1)
y=movie_df['Rating']
```

```python
# Splitting the data into train and test data
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.25, random_state=42)
```

Since the aim is to predict the ratings i am going to use foolowing regression models:

    1. RandomForestRegressor()
    2. LinearRegression()
    3. Support Vector Regression

```python
models={
    'Linear Regression':LinearRegression(),
    'Random Forest Regressor':RandomForestRegressor(),
    'Support vector Regressor':SVR()
}
```

```python
# function to fit and evaluate the models
def fit_and_evaluate(models,X_train,X_test,y_train,y_test):
    mse={}
    mae={}
    r2_scores={}
    for name,model in models.items():
#         Fit the model
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        print(f"Evaluating {name}")
        print('Mean Squared Error : ',mean_absolute_error(y_test, y_pred))
        print('Mean Absolute Error : ',mean_squared_error(y_test,y_pred))
        print('R2 Score',r2_score(y_test,y_pred))
    return fit_and_evaluate
```

```python
fit_and_evaluate=fit_and_evaluate(models=models, X_train=X_train,X_test=X_test,y_train=y_train ,y_test=y_test)
```
